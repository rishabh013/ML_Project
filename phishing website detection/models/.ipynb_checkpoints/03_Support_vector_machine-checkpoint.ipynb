```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.svm import SVC
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
path = os.getcwd()
```

```python
#importing the dataset
dataset = pd.read_csv(f"{path}/phishcoop.csv")
dataset
```

```python
dataset = dataset.drop('id', 1) #removing unwanted column

x = dataset.iloc[ : , :-1].values
y = dataset.iloc[:, -1:].values
```

```python
#spliting the dataset into training set and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.25, random_state =0 )
```

```python
#applying grid search to find best performing parameters 
from sklearn.model_selection import GridSearchCV
parameters = [{'C':[1, 10, 100, 1000], 'gamma': [ 0.1, 0.2,0.3, 0.5]}]
grid_search = GridSearchCV(SVC(kernel='rbf' ),  parameters,cv =5, n_jobs= -1)
grid_search.fit(x_train, y_train)
```

```python
#printing best parameters 
print("Best Accurancy =" +str( grid_search.best_score_))
print("best parameters =" + str(grid_search.best_params_)) 
```

```python
#fitting kernel SVM  with best parameters calculated 

classifier = SVC(C=100, kernel = 'rbf', gamma = 0.1 , random_state = 0)
classifier.fit(x_train, y_train)
```

```python
#predicting the tests set result
y_pred = classifier.predict(x_test)
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
import matplotlib.pyplot as plt
from sklearn import metrics
```

```python
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = [False, True])
cm_display.plot()
plt.show()
```
